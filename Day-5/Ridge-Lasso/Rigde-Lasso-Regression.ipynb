    📦 Import Libraries & Load Dataset 🗄️

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = load_diabetes()
print(data.DESCR)
```

<h1 style="
    color: #2c3e50;
    font-size: 32px;
    font-family: 'Geneva';
    border: 3px solid #e74c3c;
    border-radius: 10px;
    display: inline-block;
    padding: 15px 30px;
    text-align: center;
    box-shadow: 2px 4px 10px rgba(0,0,0,0.1);
">
    🚂 Train - Test Split 🔀
</h1>

```python
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.20, random_state = 42)
```

<h1 style="
    color: #2c3e50;
    font-size: 32px;
    font-family: 'Geneva';
    border: 3px solid #e74c3c;
    border-radius: 10px;
    display: inline-block;
    padding: 15px 30px;
    text-align: center;
    box-shadow: 2px 4px 10px rgba(0,0,0,0.1);
">
    🪢 Lasso Regression 🔗
</h1>

```python
alphas = [0.0001, 0.001, 0.01, 0.1, 1, 10]

for alpha in alphas:
    r = Ridge(alpha=alpha)
    r.fit(X_train, y_train)
    
    y_pred = r.predict(X_test)
    mse_ridge = mean_squared_error(y_test, y_pred)
    
    print(f"Ridge Train R²: {r.score(X_train, y_train)}")
    print(f"Ridge Test R²: {r.score(X_test, y_test)}")
    print(f"Ridge Test MSE: {mse_ridge}")
    print("\n")
```

<h1 style="
    color: #2c3e50;
    font-size: 32px;
    font-family: 'Geneva';
    border: 3px solid #e74c3c;
    border-radius: 10px;

">
    🏔️ Ridge Regression ⛰️
</h1>

```python
alphas = [0.0001, 0.001, 0.01, 0.1, 1, 10]

for alpha in alphas:
    l = Lasso(alpha = alpha)
    l.fit(X_train, y_train)
    
    pred = l.predict(X_test)
    mse_lasso = mean_squared_error(y_test, pred)

    print(f"Ridge Train R²: {l.score(X_train, y_train)}")
    print(f"Ridge Test R²: {l.score(X_test, y_test)}")
    print(f"Ridge Test MSE: {mse_lasso}")
    print("\n")
```
