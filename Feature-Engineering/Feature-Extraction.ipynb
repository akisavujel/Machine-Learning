# Feature Extration

#### using the already existing data on the dataset to create new more meaning ful data to help predict better model

```python
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.dtypes
```

```python
bins = [10,30,50,70]
labels = ["young","mid","senior"]
```

#### a new column is cut from column age where bins is seperated values to group and label is labelling them 10, 30, 50, 70 are the values where 10 is young, 30 and 50 is mid and 70 is senior

```python
df["age_group"] = pd.cut(df["age"], bins = bins, labels = labels)
```

```python
df[["age","age_group"]].head()
```

```python
df['signup_date'] = pd.to_datetime(df['signup_date'])
df['last_purchase_date'] = pd.to_datetime(df['last_purchase_date'])
```

```python
df.dtypes
```

#### converted into dateandtime datatype and extract year and month from dataset to create new columns and use todays date and signin date to find the total date for sign in same with last purchase this shows how long the customer hasnt purchased anything

```python
df['signup_year'] = df['signup_date'].dt.year
df['signup_month'] = df['signup_date'].dt.month

df['date_since_signin'] = (pd.Timestamp.today() - df['signup_date']).dt.days
df['date_since_lastpur'] = (pd.Timestamp.today() - df['last_purchase_date']).dt.days
print(df['date_since_signin'].head(3))
print(df['date_since_lastpur'].head(3))
```

#### apply lamba function from start to end split the values and 0 means select the first word and label as name and -1 is label for last values which is now name lastname and length of name

```python
df['first_name'] = df['name'].apply(lambda x:x.split()[0])
df['last_name'] = df['name'].apply(lambda x:x.split()[-1])
df['length_name'] = df['name'].apply(len)
print(df['first_name'].head(3))
print(df['last_name'].head(3))
print(df['length_name'].head(3))
```

#### if the total spend of custon=mer is more then 5000 then theu are high spender which is 1 and if its low it will be 0|

```python
df['high_spender'] = (df['total_spent'] > 5000).astype(int)
print(df['high_spender'])
```
