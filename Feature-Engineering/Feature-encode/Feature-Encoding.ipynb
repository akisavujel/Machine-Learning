```python
import pandas as pd

df = pd.read_csv('featureeng.csv')

df.head()
```

# Feature Encoding

Label Encoding: converts categorical labels into numbers.

Example: "Yes" → 1, "No" → 0

You now have a numeric column Promotion Enc that models can understand.

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

label_encoder = LabelEncoder()

df_encoded["Promotion Enc"] = label_encoder.fit_transform(df_encoded["Promotion"])

df_encoded.head()
```

# Label Encoding

Instead of using LabelEncoder, you manually assign numbers to categories.

Example:

High School → 0

Bachelor's → 1

Master’s → 2

PhD → 3

This is useful when the categories have an inherent order (ordinal features).

```python
education_encoded_dict = {'High School':0,"Bachelor's":1,"Master's":2,'PhD':3}

df_encoded["Education_Enc"] = df_encoded["Education"].map(education_encoded_dict)

df_encoded.head()
```

# One Hot Encoding

One-Hot Encoding → creates binary columns for each category.

Example: If City has ['Kathmandu', 'Pokhara'] → you get:

City_Kathmandu → 1 if city is Kathmandu, 0 otherwise

City_Pokhara → 1 if city is Pokhara, 0 otherwise

This avoids giving the model the wrong idea of order (unlike label encoding).

```python
df_encoded = pd.get_dummies(df_encoded, columns = ["City"], prefix = ["City"])

df_encoded.head()
```

# Frequency Encoding

Frequency Encoding → replaces each category with how often it appears in the dataset.

Example:

If "Data Analyst" appears 10 times, "Manager" 5 times:

"Data Analyst" → 10

"Manager" → 5

Useful for high-cardinality features (many unique values) to give the model some idea of importance.

```python
job_title_counts = df_encoded["JobTitle"].value_counts().to_dict()

df_encoded["JobTitle_Enc"] = df_encoded["JobTitle"].map(job_title_counts)

df_encoded.head()
```

# Target Encoding

Target Encoding → replaces a category with average of the target variable for that category.

Example:

Suppose Promotion Enc (Yes=1, No=0) looks like this:

High School: 0.2  # 20% got promoted
Bachelor's: 0.5
Master's: 0.6
PhD: 0.8

Then Education_Target_Enc column will have these numbers instead of strings.

```python
education_target = df_encoded.groupby("Education")["Promotion Enc"].mean().to_dict()

df_encoded["Education_Target_Enc"] = df_encoded["Education"].map(education_target)

df_encoded.head()
```
