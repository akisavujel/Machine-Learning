### Importing Scalers

We import two scaling techniques from `sklearn.preprocessing`:

- **StandardScaler** – Standardizes features by removing the mean and scaling to unit variance.
- **MinMaxScaler** – Scales features to a fixed range (usually 0 to 1).

````python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
````

````python
standardscaler = StandardScaler()
````

````python
df['salarystdscaler'] = standardscaler.fit_transform(df[['Salary']])
````

````python
minmax = MinMaxScaler()
````

````python
df['salaryminmax'] = minmax.fit_transform(df[['Salary']])
````
